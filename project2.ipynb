```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns

df = pd.read_csv('data/unemp_rate_1_oecd.csv')
df
```

```python
df = df.loc[(df['LOCATION'] == 'CAN')|(df['LOCATION'] == 'USA')]
df.TIME = pd.to_datetime(df.TIME)
df
```

```python
f, (ax1, ax2, ax3) = plt.subplots(
    1, 3, figsize=(12, 4), sharey=True)
sns.barplot(x='LOCATION', y='Value', data=df, ax=ax1)
sns.violinplot(x='LOCATION', y='Value', data=df, ax=ax2)
sns.swarmplot(x='LOCATION', y='Value', data=df, ax=ax3)
```
