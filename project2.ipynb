```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns

df = pd.read_csv('data/unemp_rate_1_oecd.csv')
df
```

```python
df['LOCATION'].unique()
```

```python
# df = df.loc[(df['LOCATION'] == 'CAN')|(df['LOCATION'] == 'USA')]
# # df.TIME = pd.to_datetime(df.TIME)
# df

# data
df_can = df.loc[(df['LOCATION'] == 'CAN')]
df_usa = df.loc[(df['LOCATION'] == 'USA')]

df_swe = df.loc[(df['LOCATION'] == 'SWE')]
df_dnk = df.loc[(df['LOCATION'] == 'DNK')]

# df_can['TIME'] = pd.to_datetime(df_can['TIME'])


fig, ax = plt.subplots(figsize=(20,15))
ax.set_xlabel('Year', fontsize=15)
ax.set_ylabel('Unemployment %', fontsize=15)
ax.set_title('Unemployment Rates', fontsize=20)

# 08 great recession
gr_start = pd.to_datetime('2007-12-01')
gr_end = pd.to_datetime('2009-07-01')

# covid recession
cr_start = pd.to_datetime('2020-02-01')
cr_end = pd.to_datetime('2020-11-01')

# markers
mark = np.array([gr_start,gr_end,cr_start,cr_end])

'''
https://fas.org/sgp/crs/misc/R46554.pdf

During the Great Recession, the unemployment rate increased from 5% in December 2007 (the
start of the recession) to 9.5% in June 2009 (the end of the recession) (see Figure 2). The
unemployment rate peaked at 10% in October 2009, four months after the recession officially
concluded. In the current recession, the unemployment rate increased from 3.5% in February
2020 to 4.4% in March 2020, peaked8 at 14.7% in April, and then fell to 6.7% in November. The
peak represents the quickest month-over-month increase in unemployment rates and the highest
overall unemployment rate since the CPS data started being collected in 1948.
'''



ax.plot_date(pd.to_datetime(df_can.TIME), df_can.Value, '-', color='C3', 
             lw=5, alpha=1, label='CAN')
ax.plot_date(pd.to_datetime(df_usa.TIME), df_usa.Value, '-', color='C9',
             lw=5, alpha=0.6, label='USA')

ax.legend(loc='upper left', fontsize=15)
# ax.plot_date(pd.to_datetime(df_swe.TIME), df_swe.Value, '-', color='C4', #marker='*',
#              lw=5, alpha=0.6, label='SWE', markevery=mark.all())
# ax.plot_date(pd.to_datetime(df_dnk.TIME), df_dnk.Value, '-', color='C5',#marker='o',
#              lw=5, alpha=0.6, label='DNK', markevery=mark.all())

# recession areas
ax.axvspan(gr_start, gr_end, color='C6', alpha=0.5)
ax.axvspan(cr_start, cr_end, color='C8', alpha=0.6)
```

```python
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

labels = df4_can.TIME
len(labels)

can_15_24 = df4_can.Value.round(1).array
usa_15_24 = df4_usa.Value.round(1).array
can_25_74 = df5_can.Value.round(1).array
usa_25_74 = df5_usa.Value.round(1).array

print(usa_25_74)


# x = np.arange(len(labels))  # the label locations
print(x)
width = 0.22  # the width of the bars

r1 = np.arange(len(labels))
r2 = [x + width for x in r1]
r3 = [x + width for x in r2]
r4 = [x + width for x in r3]


fig, ax = plt.subplots(figsize = (20,12))
rects1 = ax.bar(r1, can_15_24, width, label='can, 15-24', color='C3')
rects2 = ax.bar(r2, usa_15_24, width, label='usa, 15-24', color='C9')
rects3 = ax.bar(r3,can_25_74, width, label='can, 25-74', color='C1')
rects4 = ax.bar(r4,usa_25_74, width, label='usa, 25-74', color='C4')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Unemployment %', fontsize=25)
ax.set_xlabel('Time (YYYY/MM)', fontsize=25)
ax.set_title('Unemployment % by age group', fontsize=25)
ax.set_xticks([r + 1.5 * width for r in range(len(can_15_24))])
ax.set_xticklabels(labels)
ax.legend(fontsize=25)
ax.legend(loc='upper left', fontsize=20)


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)

fig.tight_layout()

plt.show()
```
